# Zomato Data Analysis Using Python


# Import necessary Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Create the Data Frame

```python
dataframe = pd.read_csv("Zomato_data.csv")
dataframe.head()
```

```python
dataframe.tail()
```

```python
dataframe.shape
```

# let's convert the data type of the "rate" column to float and remove the denominator

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe.head()
```

# Summary of dataframe

```python
dataframe.info()
```

# Conclusion- There is no NULL value in dataframe

# Type of Restaurant

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

# Conclusion- The majority of the restaurants fall into the dining category.

# Dining restaurants are preferred by larger number of individuals.

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of Restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

# Conclusion- Dining restaurants received maximum votes.

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```

# Conclusion- The majority of restaurants received ratings ranging from 3.5 to 4.

# Average order spending by couples

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# Conclusion- The majority of couples prefer restaurants with an approximate cost of Rs.300

# Which mode receives maximum ratings

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

# Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings.

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("online_order")
plt.ylabel("listed_in(type)")
plt.show()
```

# Conclusion- Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at dining restaurants, but prefer online ordering at cafes.
